```python
# Define a class named Person
class Person:
    # Constructor to initialize attributes
    def __init__(self, name, age):
        self.name = name  # Attribute: name
        self.age = age    # Attribute: age

    # Method to introduce the person
    def introduce(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Creating objects (instances) of the Person class
person1 = Person("Ishrath", 20)
person2 = Person("Bob", 25)

# Accessing attributes and calling methods of objects
print(person1.name)     # Output: Alice
print(person2.age)      # Output: 25
person1.introduce()     # Output: Hello, my name is Alice and I am 30 years old.
person2.introduce()     # Output: Hello, my name is Bob and I am 25 years old.
```

```python
# Define a class named Person
class Person:
    # Constructor to initialize attributes (name and age)
    def __init__(self, name, age):
        self.name = name    # Attribute: name
        self.age = age      # Attribute: age

    # Method to return a string representation of the object
    def __str__(self):
        return f"{self.name} ({self.age})"

# Create an instance (object) of the Person class with name "John" and age 36
p1 = Person("John", 36)

# Print the string representation of the object p1, which calls the __str__ method
print(p1)
```

Python INHERITANCE

```python
class Person:                          # Parent class
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):                 # Child class, inherits the properties of parent(Person) class
  def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)

x = Student("Mike", "Olsen")
x.printname()
```

Exception Handling

The try block lets you test a block of code for errors.

The except block lets you handle the error.

The else block lets you execute code when there is no error.

The finally block lets you execute code, regardless of the result of the try- and except blocks.

```python
#The try block will generate an error, because X is not defined:

try:
  print(X)
except:
  print("An exception occurred")
```

```python
#The try block will generate an output, because c is defined:
a=3
b=6
c=a+b
try:
  print(c)
except:
  print("An exception occurred")
```

```python
try:
  print(p)
except NameError:
  print("Variable p is not defined")
except:
  print("Something else went wrong")
```

```python
try:
  print("Hello")
except:
  print("Something went wrong")
else:                               # Using else statement here
  print("Nothing went wrong")
```

```python
#The finally block gets executed no matter if the try block raises any errors or not:

try:
  print(q)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
```
